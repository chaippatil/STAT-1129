```python
# Project 2: American Flag 

import turtle
import time

def main():

# background
    display = turtle.getscreen()
    display.title("American Flag")
    display.bgcolor("white")

# turtle object and speed

    t = turtle.Turtle()
    t.speed(0)
    t.penup()

# flag height and width (wikipedia mentioned a 1:1.9 ratio for the height:width, therefore, 250:475 is the same ratio)
    flag_ht = 250
    flag_wth = 475

# starting points of the flag on display plane
    x1 = -250
    y1 = 120

# red and white strip measurements
    stripe_ht = flag_ht/13
    stripe_wdt = flag_wth

# star size
    star_size = 12

    def get_color(color):
        colors = {
            'red': (255, 0, 0),
            'blue': (0, 0, 255),
            'white': (255, 255, 255)}
        return colors.get(color, (0, 0, 0))
        r, g, b = get_color("red")
        r, g, b = get_color("blue")
        r, g, b = get_color("white")

# draw flag
    def draw_flag(x, y, height, width, color):
        t.goto(x,y)
        t.pendown()
        t.color(color)
        t.begin_fill()
        t.forward(width)
        t.right(90)
        t.forward(height)
        t.right(90)
        t.forward(width)
        t.right(90)
        t.forward(height)
        t.right(90)
        t.end_fill()
        t.penup()

# draw stars of flag
    def draw_star(x,y,color,length) :
        t.goto(x,y)
        t.setheading(0)
        t.pendown()
        t.begin_fill()
        t.color(color)
        for turn in range(0,5) :
            t.forward(length)
            t.right(144)
            t.forward(length)
            t.right(144)
        t.end_fill()
        t.penup()


# draw stripes of flag
    def draw_stripes():
        x = x1
        y = y1
    # 6 red and 7 white strips
        for stripe in range(0,6):
            for color in ["red", "white"]:
                draw_flag(x, y, stripe_ht, stripe_wdt, color)
                y = y - stripe_ht

    # last red stripe
        draw_flag(x, y, stripe_ht, stripe_wdt, 'red')
        y = y - stripe_ht


# blue rectangle in flag with stars
    def draw_rectangle():
        rectangle_ht = (7/13) * flag_ht
        rectangle_wdt = (0.76) * flag_ht
        draw_flag(x1, y1, rectangle_ht, rectangle_wdt, 'blue')

# drawing the stars across rows of the blue rectangle
    def draw_starrow():
        dist_of_stars = 30
        dist_bet_lines = stripe_ht + 6
        y = 112
    # create 5 rows of stars
        for row in range(0,5) :
            x = -234
        # create 6 stars in each row
            for star in range (0,6) :
                draw_star(x, y, 'white', star_size)
                x = x + dist_of_stars
            y = y - dist_bet_lines

    def draw_starrow2():
        dist_of_stars = 30
        dist_bet_lines = stripe_ht + 6
        y = 100
    # create 4 rows of stars
        for row in range(0,4) :
            x = -217
        # create 5 stars in each row
            for star in range (0,5) :
                draw_star(x, y, 'white', star_size)
                x = x + dist_of_stars
            y = y - dist_bet_lines
        
# call all the functions
    draw_stripes()
    draw_rectangle()
    draw_starrow()
    draw_starrow2()

    t.hideturtle()

if __name__ == "__main__":
    main()
```

```python
# Project 2: Bonus - Japanese Flag

import turtle as t

def main():

    t.bgcolor('blue')
    t.up()
    t.left(90)
    t.forward(100)
    t.right(90)
    t.back(150)

    t.write('Japanese Flag')
    t.color('white')
    t.begin_fill()
    for _ in range(2):
        t.forward(300)
        t.right(90)
        t.forward(200)
        t.right(90)
    t.end_fill()

    t.color('red')
    t.forward(150)
    t.right(90)
    t.forward(40)
    t.left(90)
    t.begin_fill()
    t.circle(-60)
    t.end_fill()

    t.hideturtle()

if __name__ == "__main__":
    main()
```
