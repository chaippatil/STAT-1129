```python
# Question 2
import numpy as np
import pandas as pd
amazon = pd.read_csv("amazon-orders.csv")
df = pd.DataFrame(amazon)
df
```

```python
apriltotal = df.loc[0:9, ["Total Charged"]]
print(apriltotal)
apriltotallist = (36.93, 16.99, 9.99, 156.12, 14.99, 12.65, 40.63, 44.99, 28.80, 37.08)
aprilsum = sum(apriltotallist)
aprilmin = min(apriltotallist)
aprilmax = max(apriltotallist)
aprilmedian = (apriltotallist[len(apriltotallist) // 2] + apriltotallist[~len(apriltotallist) // 2]) / 2
aprilmean = sum(apriltotallist)/len(apriltotallist)
aprilvariance = sum([((x - aprilmean) ** 2) for x in apriltotallist]) / len(apriltotallist)
aprilstd = aprilvariance ** 0.5

print("The total amount of money spent on Amazon in April is", aprilsum)
print("The minimum amount of money spent on Amazon in April is", aprilmin)
print("The maximum amount of money spent on Amazon in April is", aprilmax)
print("The median amount of money spent on Amazon in April is", aprilmedian)
print("The average amount of money spent on Amazon in April is", aprilmean)
print("The standard deviation of the amount of money spent on Amazon in April is", aprilstd)
```
